```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fnIter(fn,x,delta=0.000001):
    while True:
        yield x
        prev,x = x,fn(x)
        if abs(x-prev)<delta:break
```

```python
r = 3.99
seed = 0.5

total = []
```

```python
for i,Xn in enumerate(fnIter(lambda x:x*r*(1-x),seed)):
    total.append(Xn)
    if (i == 500):
        break
```

```python
fig, ax = plt.subplots()
ax.plot(total)
```
